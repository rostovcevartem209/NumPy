```python
np.random.seed(42)

X = np.random.randn(100,3) #100 объектов 3 признака
w_true = np.array([2.5,-1.3,0.8]) #истинные веса (коэфициенты влияющие на целевую переменную)
noise = np.random.normal(0,0.5,100) #шум (помехи, наличие которых симулирует настоящие данные)

y = X@w_true+noise #целевые переменные

#связно перемешиваем данные (y[0] всё также является целевым значением x[0])
indices = np.arange(100)
X_shuffle = X[indices]
y_shuffle = y[indices]

#делим данные на тренировочные и тестовые
split_ind = 80
X_train, X_test = X_shuffle[:split_ind], X_shuffle[split_ind:]
y_train, y_test = y_shuffle[:split_ind], y_shuffle[split_ind:]

#обучение модели через lstsq (????)
w_estimated = np.linalg.lstsq(X_train,y_train,rcond=None)[0]

#предсказываем целевые значения
y_pred = X_test@w_estimated

#считаем среднюю квадратичную ошибку
mse = np.mean((y_pred-y_test)**2)
print(mse)
```
